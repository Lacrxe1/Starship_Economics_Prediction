```python
import pandas as pd
import numpy as np

df = pd.read_csv('/kaggle/input/finance-and-economics-dataset-2000-present/finance_economics_dataset.csv')

print("Start of the data")
print(df.head())

print("End of the data")
print(df.tail())
```

```python
# Check if data has blank cells
print(df.isnull().sum())
```

```python
# Linear Regression 
import statsmodels.api as sm

X = df['Crude Oil Price (USD per Barrel)']
Y = df['Inflation Rate (%)']

X = sm.add_constant(X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```
